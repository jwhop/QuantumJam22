```python
qc = QuantumCircuit(4, 4)
qc.h(0)
qc.cx(0, 1)
qc.measure([0, 1], [0, 1])
```

```python
#use a real backend
backend = provider.get_backend('simulator_statevector') #use a similar to test your circuit before using real hardware
#backend = provider.get_backend('ibmq_belem') #this can be swapped for any device you have access to under the "computre resources" section of quantum-computing.ibm.com


shots = 5000
job = execute(qc, backend, shots=shots, memory=True)

#we want to see results independently, instead of as a probability
result = job.result()
memory = result.get_memory()

print(result.get_memory())
```
